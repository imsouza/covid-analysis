# Data analysis and visualization

```python
def plot_line(title,xlabel,ylabel,x,y,dataset,hue):
    ax = sns.lineplot(x=x, y=y, hue=hue, data=dataset)
    ax.figure.set_size_inches(20,8)
    ax.set_title(title,loc='left',fontsize=24)
    ax.set_xlabel(xlabel,fontsize=16)
    ax.set_ylabel(ylabel,fontsize=16,rotation=90)
    ax = ax
    
def plot_bar(title,xlabel,ylabel,x,y,dataset):
    plt.figure(figsize=(20,8))
    ax = plt.bar(dataset[x],dataset[y])
    plt.title(title,loc='left',fontsize=24)
    plt.xlabel(xlabel,fontsize=16)
    plt.ylabel(ylabel,fontsize=16,rotation=90)
    ax = ax

def plot_decompose_3(y1,y2,y3,color1,color2,color3,label1,label2,label3):
    ax = plt.figure(figsize=(20,8))
    plt.plot(y1, color=color1, label=label1)
    plt.plot(y2, color=color2, label=label2)
    plt.plot(y3, color=color3, label=label3)
    plt.legend(fontsize=14)

def plot_decompose_4(y1,y2,y3,y4,color1,color2,color3,color4,label1,label2,label3,label4):
    ax = plt.figure(figsize=(20,8))
    plt.plot(y1, color=color1, label=label1)
    plt.plot(y2, color=color2, label=label2)
    plt.plot(y3, color=color3, label=label3)
    plt.plot(y4, color=color4, label=label4)
    plt.legend(fontsize=14)

def plot_acf_data(y1,y2,y3):
    fig = plt.figure(figsize=(10,10))
    ax1 = fig.add_subplot(3,1,1)
    fig = plot_acf(y1, ax=ax1, title='Original')
    ax2 = fig.add_subplot(3,1,2)
    fig = plot_acf(y2, ax=ax2, title='First differentiation')
    ax3 = fig.add_subplot(3,1,3)
    fig = plot_acf(y3, ax=ax3, title='Second differentiation')
    fig=fig
```

# Forecast models

## ARIMA Model (p=1,d=1,q=1)

> p: is the number of autoregressive terms

> d: is the number of nonseasonal differences needed for stationarity

> q: is the number of lagged forecast errors in the prediction equation

```python
model = ARIMA(y, order=(1,1,1))
model_fit = model.fit(disp=0)
print(model_fit.summary())
```

## SARIMA Model

```python
p = d = q = range(0, 2)
pdq = list(itertools.product(p, d, q))
```

```python
seasonal_pdq = [(x[0], x[1], x[2], 7) for x in list(itertools.product(p, d, q))]
```

### Parameter combination

```python
for i in range(8):
    print(f"SARIMAX: {pdq[0]} x {seasonal_pdq[i]}")
```

### Computing SARIMAX

```python
y = data_es['totalCases']
y
```

```python
aic_info = []

# iterate over the non-seasonal part
for param in pdq:
    # iterate over the seasonal part
    for param_seasonal in seasonal_pdq:
        mod = sm.tsa.statespace.SARIMAX(y, order=param, seasonal_order=param_seasonal)
        
        results = mod.fit()
        aic_info.append(f"SARIMA: {param} x {param_seasonal} => AIC{results.aic}:")
```

```python
aic_info
```

```python
mod = sm.tsa.statespace.SARIMAX(y, order=(1,1,1), seasonal_order=(1,1,1,7))
results = mod.fit()
```

### Diagnostics

```python
results.plot_diagnostics(figsize=(20,8))
plt.show()
```

### Setting the time interval for prediction

```python
pred = results.get_prediction(start=pd.to_datetime('2022-08-02'), dynamic=False)
```

### Confidence interval

```python
pred_ci = pred.conf_int()
pred_ci
```

### Prediction using known and non-updated data starting in the year 2022

```python
ax = y['2022':].plot(label='Observed')
pred.predicted_mean.plot(ax=ax, label='Predicted', color='r', alpha=0.7, figsize=(20,8))
ax.fill_between(pred_ci.index, pred_ci.iloc[:,0], pred_ci.iloc[:,1], color='g', alpha=0.25)
ax.set_ylabel('COVID-19 Cases', fontsize=18)
ax.set_title('Prediction using known and non-updated data starting in the year 2022', loc='left',fontsize=24)
plt.legend()
plt.show()
```

### Prediction using unknown and non-updated data (120 days)

```python
pred_gf = results.get_forecast(steps=120)
pred_ci = pred_gf.conf_int()

ax = y.plot(label='Observed', figsize=(20,8))
pred_gf.predicted_mean.plot(ax=ax, label='Predicted', color='r', alpha=0.7)
ax.fill_between(pred_ci.index, pred_ci.iloc[:,0], pred_ci.iloc[:,1], color='g', alpha=0.1)
ax.set_xlabel('Date', fontsize=18)
ax.set_ylabel('COVID-19 Cases', fontsize=18)
ax.set_title('Prediction using unknown and non-updated data (120 days)', loc='left',fontsize=24)
plt.legend()
plt.show()
```
